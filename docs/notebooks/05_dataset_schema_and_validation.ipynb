# Dataset Schema and Validation


This notebook shows how to build datasets from arrays and run validation.


```python
import numpy as np
from crl.data.datasets import LoggedBanditDataset, TrajectoryDataset

bandit = LoggedBanditDataset.from_numpy(
    contexts=np.zeros((5, 2)),
    actions=np.array([0, 1, 0, 1, 0]),
    rewards=np.ones(5),
    behavior_action_probs=np.full(5, 0.5),
)
bandit.describe()

traj = TrajectoryDataset.from_numpy(
    observations=np.zeros((2, 3, 1)),
    actions=np.zeros((2, 3), dtype=int),
    rewards=np.ones((2, 3)),
    next_observations=np.zeros((2, 3, 1)),
    discount=0.9,
    action_space_n=2,
)
traj.describe()
```
